```python
categorical_columns = dataset.select_dtypes(['object']).columns.tolist()
numeric_columns = dataset.select_dtypes(['int64', 'float64']).columns.tolist()
# print(categorical_columns)
# print(numeric_columns)
```

```python
na_series = dataset[numeric_columns].isnull().sum() > 0
num_missing_features = na_series[na_series == True].index.tolist()
used_num_features = list(set(numeric_columns).difference(set(num_missing_features)))
data = dataset[used_num_features]
size = data.shape[1]
data_corr = data.corr()

threshold = 0.5

# List of pairs along with correlation above threshold
corr_list = []

for i in range(0,size): #for 'size' features
    for j in range(i+1,size): #avoid repetition
        if (data_corr.iloc[i,j] >= threshold and data_corr.iloc[i,j] < 1) or (data_corr.iloc[i,j] < 0 and data_corr.iloc[i,j] <= -threshold):
            corr_list.append([data_corr.iloc[i,j],i,j]) #store correlation and columns index

s_corr_list = sorted(corr_list,key=lambda x: -abs(x[0]))

for v,i,j in s_corr_list:
    print ("%s and %s = %.2f" % (used_num_features[i],used_num_features[j],v))
```

```python
#Print values highly correlated with sale price
for v,i,j in s_corr_list:
    if(used_num_features[i] == 'SalePrice' or used_num_features[j]=='SalePrice'):
      print ("%s and %s = %.2f" % (used_num_features[i],used_num_features[j],v))
```
